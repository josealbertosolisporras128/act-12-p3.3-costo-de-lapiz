1. X --> cantidad de lapices
2. PAG --> pago

## formulas
* $PAG = X * 0.85$
* $PAG = X * 0.90$

## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/josealbertosolisporras128/act-12-p3.3-costo-de-lapiz/refs/heads/main/DF%20if%20else%20lapices.png" width="300" height="500" alt="Descripción de la imagen">


```python
# para cuando es falso (no)
# Calculando el pago de lápices usando if else
# datos de entrada X (cantidad), PAG (pago)
# si es verdadero (X >= 1000) costo de 0.85
# si es falso costo de 0.90

X = int(input("Ingrese la cantidad de lápices: "))

if X >= 1000:
    PAG = X * 0.85
    print("El pago que se tiene que realizar es: ", PAG)
    print("se aplico el descuento")
else:
    PAG = X * 0.90
    print("El pago que se tiene que realizar es: ", PAG)
    print("no se aplico el descuento")

print("Fin del programa")
print("programa realizado por jose solis NL = 51")
```

```python
# para cuando es verdadero (si)
# Calculando el pago de lápices usando if else
# datos de entrada X (cantidad), PAG (pago)
# si es verdadero (X >= 1000) costo de 0.85
# si es falso costo de 0.90

X = int(input("Ingrese la cantidad de lápices: "))

if X >= 1000:
    PAG = X * 0.85
    print("El pago que se tiene que realizar es: ", PAG)
    print("se aplico el descuento")
else:
    PAG = X * 0.90
    print("El pago que se tiene que realizar es: ", PAG)
    print("no se aplico el descuento")

print("Fin del programa")
print("programa realizado por jose solis NL = 51")
```
